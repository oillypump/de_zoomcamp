## Answer_No_3

```python
total_rows = 0
taxi = "yellow"
year = 2020

for month in range(1, 13):
    file_name = f"{taxi}_tripdata_{year}-{month:02d}.csv.gz"
    url = f"https://github.com/DataTalksClub/nyc-tlc-data/releases/download/{taxi}/{file_name}"
    print(f"count file : {month:02d}...")
    df = pd.read_csv(url, low_memory=False)

    total_rows+=len(df)

print(f"{total_rows}")
```
